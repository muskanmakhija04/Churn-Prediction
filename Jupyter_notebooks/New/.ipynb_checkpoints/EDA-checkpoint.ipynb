```python
# converting 'Total Charges' object data type to numeric
df['Total_Charges'] = pd.to_numeric(df['Total_Charges'], errors='coerce')
```

```python
print(df['Tenure_Months'].max())
```

```python
# we can group the tenure in bins of 12 months

labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]
labels
```

```python
df['tenure_group'] = pd.cut(df['Tenure_Months'], range(1, 80, 12), right=False, labels=labels)
```

```python
df['tenure_group'].value_counts()
```

```python
#drop column customerID and tenure months
df.drop(columns= ['Tenure_Months'], axis=1, inplace=True)
df.head()
```

```python
for i, predictor in enumerate(df.drop(columns=['City','Zip_Code','Latitude','Longitude','Churn_Value','Total_Charges', 'Monthly_Charges'])):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Churn_Value')
```
